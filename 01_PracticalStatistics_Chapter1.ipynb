## Konsep Utama

- Median
- Outlier

## Penjelasan
EDA digunakan untuk memahami pola data. Mean bisa dipengaruhi outlier, sedangkan median lebih stabil.

```python
import numpy as np
import matplotlib.pyplot as plt

data = [10, 12, 14, 15, 18, 100]

mean = np.mean(data)
median = np.median(data)

print("Mean:", mean)
print("Median:", median)

plt.hist(data, bins=5)
plt.title("Histogram Data")
plt.show()

plt.boxplot(data)
plt.title("Boxplot (Outlier terlihat)")
plt.show()
```
