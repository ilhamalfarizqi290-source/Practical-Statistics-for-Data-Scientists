## Konsep Utama

- Similarity

## Penjelasan

Clustering digunakan untuk mengelompokkan data berdasarkan kemiripan.

Berbeda dengan supervised learning, metode ini tidak membutuhkan label.

Digunakan dalam segmentasi pelanggan dan analisis pola data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
X = np.random.rand(100, 2)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.title("KMeans Clustering")
plt.show()
```
