# Chapter 4 - Sampling & Data Bias

## Summary
Chapter ini membahas bagaimana data diambil (sampling) dan bagaimana bias dapat mempengaruhi hasil analisis.

## Konsep Utama
- Sampling
- Random Sampling
- Bias
- Sampling Error

## Penjelasan

Sampling adalah proses mengambil sebagian data dari populasi untuk dianalisis. Dalam data science, kita jarang menggunakan seluruh data karena keterbatasan waktu dan resource.

Namun, jika sampling tidak dilakukan dengan benar, maka hasil analisis bisa menjadi bias. Bias terjadi ketika sampel tidak mewakili populasi sebenarnya.

Sampling error adalah perbedaan antara hasil sampel dan populasi asli. Semakin besar jumlah sampel, biasanya semakin kecil error-nya.

Konsep ini penting karena semua model machine learning bergantung pada kualitas data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# membuat populasi
population = np.random.normal(50, 10, 10000)

# mengambil sampel
sample = np.random.choice(population, size=100)

print("Population mean:", np.mean(population))
print("Sample mean:", np.mean(sample))
```

```python
plt.hist(population, bins=50, alpha=0.5, label='Population')
plt.hist(sample, bins=20, alpha=0.7, label='Sample')
plt.legend()
plt.title("Sampling vs Population")
plt.show()
```

```python
# bandingkan beberapa sample
means = [np.mean(np.random.choice(population, size=50)) for _ in range(100)]

plt.plot(means)
plt.title("Variasi Sample Mean")
plt.show()
```
