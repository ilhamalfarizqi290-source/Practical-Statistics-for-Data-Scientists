## Penjelasan

Regression digunakan untuk memahami hubungan antara variabel independen dan dependen.

Dalam linear regression, kita mencari garis terbaik yang menggambarkan hubungan tersebut.

Konsep ini sangat penting dalam machine learning karena digunakan untuk prediksi.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10)
y = 2*x + np.random.randn(10)
```

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
coef = np.polyfit(x, y, 1)
line = coef[0]*x + coef[1]

plt.scatter(x, y)
plt.plot(x, line, color='red')
plt.title("Linear Regression")
plt.show()
```
