```python
import numpy as np
import matplotlib.pyplot as plt

trials = 1000
coin = np.random.choice([0,1], size=trials)

print("Probability Head:", np.mean(coin))

plt.bar(["Head","Tail"], [np.sum(coin), trials-np.sum(coin)])
plt.title("Coin Toss")
plt.show()

cumulative = np.cumsum(coin) / np.arange(1, trials+1)

plt.plot(cumulative)
plt.title("Cumulative Probability")
plt.show()
```
