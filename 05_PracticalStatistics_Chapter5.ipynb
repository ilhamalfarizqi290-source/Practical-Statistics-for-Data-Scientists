## Penjelasan

Statistical inference digunakan untuk memperkirakan parameter populasi dari sampel.

Confidence interval memberikan rentang nilai yang kemungkinan besar mengandung nilai sebenarnya.
Misalnya, interval 95% berarti kita cukup yakin bahwa nilai sebenarnya berada di dalam interval tersebut.

Konsep ini penting karena dalam praktik, kita jarang memiliki seluruh data populasi.

```python
import numpy as np

data = np.random.normal(100, 15, 200)

mean = np.mean(data)
std = np.std(data)

lower = mean - 1.96 * std/np.sqrt(len(data))
upper = mean + 1.96 * std/np.sqrt(len(data))
```

```python
plt.hist(data, bins=20)
plt.axvline(lower, color='red', label='Lower Bound')
plt.axvline(upper, color='red', label='Upper Bound')
plt.legend()
plt.title("Confidence Interval Visualization")
plt.show()
```

```python
sample_means = [np.mean(np.random.choice(data, size=30)) for _ in range(100)]

plt.plot(sample_means)
plt.title("Sampling Distribution of Mean")
plt.show()
```
